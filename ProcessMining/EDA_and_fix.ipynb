```python
import pandas as pd
import numpy as np
from datetime import datetime
```

# Fix datatime format

```python
df = pd.read_csv('synthetic_data_v0.csv')
df.head(20)
```

```python
date_format = '%d-%m-%Y %H:%M'
df.loc[:,'StartTimestamp'] = df.loc[:,'StartTimestamp'].apply(lambda x: datetime.strptime(x, date_format))
df.loc[:,'EndTimestamp'] = df.loc[:,'EndTimestamp'].apply(lambda x: datetime.strptime(x, date_format))
```

```python
df.to_csv('synthetic_data_v0a.csv')
```

# Analyze Activity order
## Sythetic data file

```python
df2 = pd.read_csv('synthetic_data_v0a.csv')
df2 = df2.iloc[:,2:]
df2.head(5)
```

```python
len(df2.CaseId.unique())
```

```python
df2['ActivityName'].unique()
```

```python
df2.sort_values(['CaseId','StartTimestamp','EndTimestamp'],inplace=True,ignore_index=True)
df2['counter'] = int(0)
df2.head(15)
```

```python
# go through all cases and add counter 1,2,...n for all activities in the order they START
for case_id in df2.CaseId.unique():
    counter = 1
    for index, row in df2.loc[df2['CaseId'] == case_id].sort_values(['StartTimestamp','EndTimestamp']).iterrows():
        df2.iloc[index,9] = int(counter)
        counter += 1
```

```python
df2.head(15)
```

```python
pd.set_option('display.max_rows', 500)
```

```python
# shows how many times is certain activity n-th in the order for the dame case id.
# e.g. invoce entry is often first (464 times out of 500)
df2_act = df2.groupby(['ActivityName','counter'])['ActivityName'].aggregate('count')
df2_act = pd.DataFrame(df2_act)
df2_act
```

```python
df3.sort_values(['CaseId','StartTimestamp','EndTimestamp'],inplace=True,ignore_index=True)
df3['counter'] = int(0)
df3.head()
```

```python
# go through all cases and add counter 1,2,...n for all activities in the order they START
for case_id in df3.CaseId.unique():
    counter = 1
    for index, row in df3.loc[df3['CaseId'] == case_id].sort_values(['StartTimestamp','EndTimestamp']).iterrows():
        df3.iloc[index,7] = int(counter)
        counter += 1
```
