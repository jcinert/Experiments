```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
synthetic_data=pd.read_csv("synthetic_data_processed.csv")
```

```python
original_data=pd.read_csv("ydata//SampleData_AP_Refunds_Financial_EventLog .csv")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original Data
role_counts_original = original_data['Role'].value_counts()
role_counts_original.plot(kind='bar', rot=0, color='skyblue', ax=axes[0])
axes[0].set_title('Original Data - Distribution of Roles')
axes[0].set_xlabel('Role')
axes[0].set_ylabel('Count')

# Synthetic Data
role_counts_synthetic = synthetic_data['Role'].value_counts()
role_counts_synthetic.plot(kind='bar', rot=0, color='lightcoral', ax=axes[1])
axes[1].set_title('Synthetic Data - Distribution of Roles')
axes[1].set_xlabel('Role')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original Data
role_counts_original = original_data['Resource'].value_counts()
role_counts_original.plot(kind='bar', rot=0, color='skyblue', ax=axes[0])
axes[0].set_title('Original Data - Distribution of Roles')
axes[0].set_xlabel('Resource')
axes[0].set_ylabel('Count')

# Synthetic Data
role_counts_synthetic = synthetic_data['Resource'].value_counts()
role_counts_synthetic.plot(kind='bar', rot=0, color='lightcoral', ax=axes[1])
axes[1].set_title('Synthetic Data - Distribution of Roles')
axes[1].set_xlabel('Resource')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original Data
role_counts_original = original_data['ActivityName'].value_counts()
role_counts_original.plot(kind='bar', rot=0, color='skyblue', ax=axes[0])
axes[0].set_title('Original Data - Distribution of ActivityName')
axes[0].set_xlabel('ActivityName')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=90) 


# Synthetic Data
role_counts_synthetic = synthetic_data['ActivityName'].value_counts()
role_counts_synthetic.plot(kind='bar', rot=0, color='lightcoral', ax=axes[1])
axes[1].set_title('Synthetic Data - Distribution of ActivityName')
axes[1].set_xlabel('ActivityName')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
# Convert timestamps to datetime objects
original_data['StartTimestamp'] = pd.to_datetime(original_data['StartTimestamp'])
original_data['EndTimestamp'] = pd.to_datetime(original_data['EndTimestamp'])

# Temporal analysis
temporal_analysis = original_data.groupby(original_data['StartTimestamp'].dt.date)['InvoiceValue'].sum()
temporal_analysis.plot(kind='line', marker='o', linestyle='-', color='orange')
plt.title('Temporal Analysis of InvoiceValue_original')
plt.xlabel('Date')
plt.ylabel('Total InvoiceValue')
plt.show()
```

```python
original_data['StartTimestamp'] = pd.to_datetime(original_data['StartTimestamp'])
original_data['EndTimestamp'] = pd.to_datetime(original_data['EndTimestamp'])

synthetic_data['StartTimestamp'] = pd.to_datetime(synthetic_data['StartTimestamp'])
synthetic_data['EndTimestamp'] = pd.to_datetime(synthetic_data['EndTimestamp'])

# Temporal analysis for both datasets
temporal_analysis_original = original_data.groupby(original_data['StartTimestamp'].dt.date)['InvoiceValue'].sum()
temporal_analysis_synthetic = synthetic_data.groupby(synthetic_data['StartTimestamp'].dt.date)['InvoiceValue'].sum()

# Plotting both graphs on the same console
plt.figure(figsize=(10, 6))

temporal_analysis_original.plot(kind='line', marker='o', linestyle='-', color='orange', label='Original Data')
temporal_analysis_synthetic.plot(kind='line', marker='o', linestyle='-', color='blue', label='Synthetic Data')

plt.title('Temporal Analysis of InvoiceValue')
plt.xlabel('Date')
plt.ylabel('Total InvoiceValue')
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original Data
vendor_counts_original = original_data['VendorID'].value_counts()
vendor_counts_original.plot(kind='bar', rot=0, color='skyblue', ax=axes[0])
axes[0].set_title('Original Data - Distribution of Vendors')
axes[0].set_xlabel('VendorID')
axes[0].set_ylabel('Count')

# Synthetic Data
vendor_counts_synthetic = synthetic_data['VendorID'].value_counts()
vendor_counts_synthetic.plot(kind='bar', rot=0, color='lightcoral', ax=axes[1])
axes[1].set_title('Synthetic Data - Distribution of Vendors')
axes[1].set_xlabel('VendorID')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python


# Kolmogorov-Smirnov Test for Categorical Features
categorical_features = original_data_categorical.columns
for feature in categorical_features:
    ks_statistic, p_value = ks_2samp(original_data_categorical[feature], synthetic_data_categorical[feature])
    print(f'KS Statistic for {feature}: {ks_statistic}, p-value: {p_value}')

# Visual Comparison for Categorical Features
for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    plt.title(f'{feature} - Original vs. Synthetic')
    original_data[feature].value_counts().plot(kind='bar', alpha=0.5, color='blue', label='Original')
    synthetic_data[feature].value_counts().plot(kind='bar', alpha=0.5, color='orange', label='Synthetic')
    plt.legend()
    plt.show()
```
