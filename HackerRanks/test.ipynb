```python
def resolveCountries(n, astronaut):

    astr_id = list(range(n))
    astr_country = list(range(n))
    for a in astronaut:
        a.sort()
    astronaut.sort()

    for idx, astr_x in enumerate(astr_id):
        set_country(astr_x, astr_country[idx],0,astronaut,astr_country)
    return astr_id, astr_country
```

```python
def set_country(astr_id,country_id,search_id,astronaut,astr_country):
    for dupl_c in astronaut[search_id:]:
        if astr_id in dupl_c:
            astr_country[dupl_c[0]] = country_id
            astr_country[dupl_c[1]] = country_id
            set_country(dupl_c[0],country_id,search_id+1,astronaut,astr_country)
            set_country(dupl_c[1],country_id,search_id+1,astronaut,astr_country)
```

```python
def journeyToMoon(n, astronaut):
    # Write your code here
    astr_id, astr_country = resolveCountries(n, astronaut)
    
    print('---INPUT--------')
    print(f'All astronauts: {astr_id}')
    print(f'Same country: {astronaut}')
    print('---WIP--------')

    print(f'Astr countries: {astr_country}')
    n_comb = 0
    for i, a1 in enumerate(astr_id):
        for ii, a2 in enumerate(astr_id[i+1:]):
            if astr_country[i] != astr_country[ii+i+1]:
                n_comb += 1
                print(f'--team: {i}, {ii+i+1}')
            else:
                print(f'--team: {i}, {ii+i+1} - same country')


    print('---OUTPUT--------')    
    print(f'Comb: {n_comb}')
```
